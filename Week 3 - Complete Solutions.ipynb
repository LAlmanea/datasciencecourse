# Barplot

**Create a df "df_2017" where you take all value in the year 2017** Q

Tip: Use column name "year"

```python
df_2017 = df[df.year == 2017]
df_2017.shape
```

```python
df_2017 = df[df.Date.dt.year == 2017]
df_2017.shape
```

How many rows are there in df_2017?

5722

**Now Groupby "region" and "type" on df_2017. Calculate the mean of the "AveragePrice" column. Reset the index and save it in a variable called "df_2017_region_price".**

```python
df_2017_region_price = df_2017.groupby(['region', 'type']).AveragePrice.mean().reset_index()
df_2017_region_price
```

**How many rows and columns does "df_2017_region_price" have?**

108,3

```python
df_2017_region_price.sort_values(by=['type', 'AveragePrice'],ascending=True, inplace=True)
```

```python
city_name = ['NewYork', 'LosAngeles', 'Chicago','Houston', 'Denver', 'Philadelphia', 'Seattle', 'SanDiego', 'SanFrancisco', 'Seattle']
```

**Create a new df with only the cities in "city_name"**

```python
df_2017_city_price = df_2017_region_price[df_2017_region_price.region.isin(city_name)]
```

**Plot a barchart ("region" and "AveragePrice") with the new data frame. Plot only from the conventional avocados.**

```python
fig = px.bar(df_2017_city_price[df_2017_city_price.type == 'conventional'], y='region', x='AveragePrice', orientation='h')
fig.show()
```

```python
fig = px.bar(df_2017_city_price[df_2017_city_price.type == 'conventional'], x='region', y='AveragePrice', orientation='v')
fig.show()
```

**Now do the same with the organic ones**

```python
fig = px.bar(df_2017_city_price[df_2017_city_price.type == 'organic'], y='region', x='AveragePrice', orientation='h')
fig.show()
```

**Now stack the two barplots**

```python
df_2017_city_price.sort_values(by=['AveragePrice'],ascending=False, inplace=True)
fig = px.bar(df_2017_city_price, x='region', y='AveragePrice', color='type',barmode='group')
fig.show()
```
